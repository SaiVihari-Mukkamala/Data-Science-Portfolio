```python
def numeric_plots(df, column):
    fig, (ax1, ax2) = plt.subplots(1, 2)

    # Distribution plot (left)
    sns.histplot(data=df, x=column, kde=True, ax=ax1)
    ax1.set_title(f'{column} Distribution')

    # Box plot (right)
    sns.boxplot(x=column, data=df, ax=ax2)

    plt.show()
```

```python
for col in numeric_cols:
    numeric_plots(df, col)
```

```python
def outlier_check(x):
    iqr = df[x].quantile(0.75) - df[x].quantile(0.25)

    uppers = df[x].quantile(0.75) + 1.5 * iqr

    lowers = df[x].quantile(0.25) - 1.5 * iqr

    outlier = 0

    for i in df[x]:
        if outlier <= 50:
            if i > uppers or i < lowers:
                outlier += 1
        else:
            return True
```

```python
outlier_columns = []
for col in numeric_cols:
    if outlier_check(col):
        outlier_columns.append(col) 
```

These outliers are quite common in financial data.

Capping Outliers

```python
def count_outliers(data):
    # Calculate first and third quartiles
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    
    # Calculate interquartile range (IQR)
    iqr = q3 - q1
    
    # Define the outlier boundaries
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Count outliers
    outliers = (data < lower_bound) | (data > upper_bound)
    
    return np.sum(outliers)
```

```python
for i in outlier_columns:
    upper = df[i].quantile(0.90)
    lower = df[i].quantile(0.10)
    df[i] = df[i].apply(lambda x: upper if x > upper else (lower if x < lower else x))
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder

X = df.drop(columns='Credit_Score')
y = df['Credit_Score'].apply(lambda x : 1 if x == 'Good' else (2 if x == 'Standard' else 3))


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score,
    r2_score, precision_recall_curve, precision_score, recall_score, roc_curve, auc, average_precision_score)
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def pipe(s4, param_grid):
    s1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), ['Payment_of_Min_Amount', 'Credit_Mix', 'Payment_Behaviour'])
    ], remainder='passthrough')

    s2 = StandardScaler()

    s3 = SMOTE(sampling_strategy='auto', random_state=42)
    pipeline = ImbPipeline([
        ('ColumnTransformer', s1),
        ('scaler', s2),
        ('smote', s3),  # Use SMOTE directly in the imbalanced-learn Pipeline
        ('clf', s4)
    ])

    random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=param_grid, n_iter=10, cv=5, random_state=42)

    random_search.fit(x_train, y_train)

    best_model = random_search.best_estimator_

    return best_model
```

```python
def plot__(ytest, preds):
    from sklearn.metrics import precision_recall_curve, average_precision_score, roc_curve, auc

    precision = dict()
    recall = dict()
    average_precision = dict()
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(1, 4):  # Replace 3 with the number of classes in your data
        precision[i], recall[i], _ = precision_recall_curve(y_test == i, (preds == i).astype(int))
        average_precision[i] = average_precision_score(y_test == i, (preds == i).astype(int))
        
        fpr[i], tpr[i], _ = roc_curve(y_test == i, (preds == i).astype(int))
        roc_auc[i] = auc(fpr[i], tpr[i])

    import matplotlib.pyplot as plt

    plt.figure(figsize=(16, 5))

    # Plot PR curves
    plt.subplot(1, 2, 1)
    for i in range(1, 4):  # Replace 3 with the number of classes in your data
        plt.plot(recall[i], precision[i], label=f'Class {i} (AUC={average_precision[i]:.2f})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curves (Multi-Class)')
    plt.legend()

    # Plot ROC curves
    plt.subplot(1, 2, 2)
    for i in range(1, 4):  # Replace 3 with the number of classes in your data
        plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC={roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curves (Multi-Class)')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
classifiers = [
    (LogisticRegression(max_iter=1000, multi_class='auto'), {
        'clf__C': [0.01, 0.1, 1],
        'clf__solver': ['liblinear', 'saga']
    }),
    (KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto'),{
    'clf__n_neighbors': [3, 5, 7],
    'clf__weights': ['uniform', 'distance'],
    'clf__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}),
    (DecisionTreeClassifier(criterion='entropy', random_state=42), {
        'clf__max_depth': [None, 5],
        'clf__min_samples_split': [2, 5],
        'clf__min_samples_leaf': [1, 2]
    }),
    (RandomForestClassifier(random_state=42), {
        'clf__n_estimators': [50, 100],
        'clf__max_depth': [None, 5],
        'clf__min_samples_split': [2, 5],
        'clf__min_samples_leaf': [1, 2]
    }),(GradientBoostingClassifier(random_state=42), {
    'clf__n_estimators': [100, 200, 300],
    'clf__learning_rate': [0.01, 0.1, 0.5],
    'clf__max_depth': [3, 5, 7],
    'clf__min_samples_split': [2, 5, 10],
    'clf__min_samples_leaf': [1, 2, 4],
    'clf__subsample': [0.5, 0.8, 1.0],
    'clf__colsample_bytree': [0.5, 0.8, 1.0],
}),
    (XGBClassifier(objective='multi:softmax', random_state=42), {
        'clf__n_estimators': [100, 200],
        'clf__max_depth': [3, 4],
        'clf__learning_rate': [0.1, 0.01]
    })
]
```

```python
names = ['LogReg', 'KNN', 'DT', 'RForest', 'GBDT', 'XGBoost']

model_list = []
```

```python
for name, classifier in zip(names,classifiers):
    model = name
    name = pipe(classifier[0], classifier[1])

    model_list.append(name)

    preds = name.predict(x_test)
    print('---'*60)
    print(f'{model}')
    print('---'*60)
    print(classification_report(y_test, preds))
    print('\n')
    print('---'*60)
    plot__(y_test, preds)
```
